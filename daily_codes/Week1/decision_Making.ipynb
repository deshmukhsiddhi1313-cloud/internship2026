```python
# menu driven
print("welcome to the simple calculator")

num = int(input("enter the number"))

print("Menu=>\n1. Square\n2. Cube\n3. check +ve or -ve")

ch = int(input("select ur choice"))

if ch == 1:
    print(num * num)
elif ch == 2:
    print(num ** 3)
elif ch == 3:
    if num > 0:
        print("positive")
    elif num == 0:
        print("zero")
    else:
        print("negative")
else:
    print("Please visit nearest hospital to check ur eyes!!!! ")
```

```python
""" 
menu driven ->
take a string from the user
menu -> 1 upper, 2 lowercase, 3 replace a particular part of the string
for third option it should which part to replace with what
""";
```

```python
print("welcome")
text = input("enter the string")

print("menu =>\n1. upper\n2. lower\n3. replace")

ch = int(input("enter the choice"))

if ch == 1:
    print(f"{text.upper()}")
elif ch == 2:
    print(f"{text.lower()}")
elif ch == 3:
    r1 = input("which part u want to replace: ")
    r2 = input("what u want to replace it with")
    new = text.replace(r1, r2)
    print(f"result -> {new}")
```
